```python
from azureml.core import Workspace, Environment
from azureml.core.model import Model
from azureml.core.model import InferenceConfig
from azureml.core.environment import Environment
from azureml.core.conda_dependencies import CondaDependencies
from azureml.core.webservice import AciWebservice, Webservice, LocalWebservice
```

```python
ws = Workspace.from_config("./deploy_setup/config.json")
```

```python
model = Model.register(model_path = "./models/",
                       model_name = "sklearn_regression_model.pkl",
                       description = "Fortune teller model",
                       workspace = ws)
```

```python
# env = Environment(name="project_environment")

# # Specify the python packages for installing
# python_packages = ["onnxruntime", "nltk", "numpy", "azureml-defaults", "joblib", "azure-ml-api-sdk"]
# for package in python_packages:
#     env.python.conda_dependencies.add_pip_package(package)
    

# Create the environment
myenv = Environment(name="myenv")
conda_dep = CondaDependencies()

# Define the packages needed by the model and scripts
# conda_dep.add_pip_package("pydantic")
conda_dep.add_pip_package("numpy")
# conda_dep.add_pip_package("nltk")
# conda_dep.add_pip_package("pydantic")
conda_dep.add_pip_package("azureml-core")
conda_dep.add_pip_package("joblib")
conda_dep.add_pip_package("onnxruntime")
conda_dep.add_pip_package("scikit-learn")
conda_dep.add_pip_package("azure-ml-api-sdk")

# You must list azureml-defaults as a pip dependency
conda_dep.add_pip_package("azureml-defaults")

# Adds dependencies to PythonSection of myenv
myenv.python.conda_dependencies=conda_dep
```

```python
inference_config = InferenceConfig(entry_script="./score.py",environment=myenv)
```

```python
# deployment_config = AciWebservice.deploy_configuration(cpu_cores = 1, memory_gb = 4)

# For local deployment
deployment_config = LocalWebservice.deploy_configuration(port=6789)
```

```python
service = Model.deploy(workspace = ws, name = "lakpa", models = [model], inference_config = inference_config, deployment_config = deployment_config)
service.wait_for_deployment(show_output = True)
print(service.get_logs())
print(service.state)
print("scoring URI: ",service.scoring_uri)
```
